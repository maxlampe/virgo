# Plotting raw simulation data

```python
from virgo.cluster import VirgoCluster

%load_ext autoreload
%autoreload 2

%matplotlib inline
```

```python
snaps = [
    "/home/max/Software/virgo/data/VIRGO/snap_800",
    "/home/max/Software/virgo/data/VIRGO/snap_850",
    "/home/max/Software/virgo/data/VIRGO/snap_900",
]
for filebase in snaps:
    virgo_cluster = VirgoCluster(file_name=filebase, io_mode=1)
    virgo_cluster.plot_raw_hists(bins=200, plot_range=[[2000., 8000.], [-6000., 1000.], [-3000., 6000.]])
```
